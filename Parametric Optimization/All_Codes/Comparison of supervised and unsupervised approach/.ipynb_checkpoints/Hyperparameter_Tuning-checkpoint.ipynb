```python
X = ab
y = xyz

from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X,y, test_size=0.3, random_state=12)
```

```python
def objective(trial):
    parameter = {
        'n_estimators': trial.suggest_int('estimators', 0,500),
        'max_depth': trial.suggest_int('depth', 3,5),
        'learning_rate': trial.suggest_loguniform('ourlearning_rate', 0.05, 0.5),
        'nthread': -1
        }
    activation = UN.unsupervised_NN.activation_func(params['activation'])
    last_activation = UN.unsupervised_NN.activation_func('sigmoid')
    model = UN.unsupervised_NN
    
    optimizer = th.optim.Adam(model.parameter(), lr=params['learning_rate'])
    loss_fn = UN.penalty_function(xyz,ab)
```

```python
num_epochs = 500

for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()
    predicted_outputs_train = model.forward(X_train)
    loss_tr = loss_fn(predicted_outputs_train, y_train)
    loss_tr.backward()
    optimizer.step()
    
    model.eval()
    with th.no_grad():
        predicted_outputs_valid = model.forward(X_valid)
        loss_vl = loss_fn(predicted_outputs_valid, y_valid)
        loss_valid = loss_vl.item()
        
return loss_valid
```

```python
study = optuna.create_study(direction='minimize', sampler=optuna.samplers.TPESampler())
study.optimize(objective, n_trials=40)

best_trial = study.best_trial
for key,value in best_trial.params.items():
    print('{}:{}'.format(key,value))
```
